```python
from konlpy.tag import Okt
from collections import Counter
import csv
```

```python
okt = Okt()
```

```python
with open("F:/eunjin/NLP/data/newslist.csv", "r", encoding='utf-8') as f:
    csvr = csv.reader(f)
    with open("news_result_x1000.csv", "a", newline='', encoding='utf-8') as f2:
        for news in csvr: # x1000 / 
            noun = okt.nouns(str(news))

            for i, j in enumerate(noun):
                if len(j)<2:
                    noun.pop(i)
                elif (j=="기자" or j=="연합뉴스"):
                    noun.pop(i)

            count = Counter(noun)
            noun_list = count.most_common(10)
         
            csvw = csv.writer(f2)
            for i in noun_list:
                csvw.writerow(i)
```
